```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://loyolagreyhounds.com/sports/womens-basketball/stats/2022-23'
path_years = 'https://loyolagreyhounds.com'
path_matchs = 'https://loyolagreyhounds.com/sports/womens-basketball/stats/2022-23/niagara/boxscore/'

data_list_gbg = []

response = requests.get(url)
soup = BeautifulSoup(response.content,'html')
main = soup.find('main',class_='main-content-placeholder')
table_seasons = main.find('select',id='ctl00_cplhMainContent_seasons_ddl')
data_years = table_seasons.find_all('option',value=True)
years = [path_years+data['value'] for data in data_years[1:]]

for year in years:
    response = requests.get(year)
    soup = BeautifulSoup(response.content,'html')
    print(f'Successfully received {year}')
    date = year.split("stats/",1)[1]
    
    game_team = soup.find('section',{'id':'game-team'})
    table = game_team.find('table',{'class':'sidearm-table'})
    
    table_body = table.find('tbody')
    match_rows = table_body.find_all('tr')
    
    for row in match_rows:
        row_data = row.find_all('td')
        individual_row_data = [data.text.strip() for data in row_data]

        link = row_data[0].find('a')['href']
        id = link.split('=')[1].split('&')[0]
        individual_row_data.extend([(path_matchs+id),date])
        data_list_gbg.append(individual_row_data)

headers_row = table.find('thead',{'class':'sidearm-primary'})
headers = headers_row.find_all('th')

columns_gbg = [column.text.strip() for column in headers]
columns_gbg.append('Link')
df_game_by_game = pd.DataFrame(data_list_gbg,columns = columns_gbg+['Date'])
    
```

```python
df_game_by_game.to_csv('all_game_by_game.csv',index=False)
df_game_by_game
```
