```python
path = 'https://loyolagreyhounds.com/sports/womens-basketball/stats/2022-23/niagara/boxscore/'
```

```python
url = "https://loyolagreyhounds.com/sports/womens-basketball/stats#game"
response = requests.get(url)

if response.status_code == 200:
    html_content = response.content
    soup = BeautifulSoup(html_content,'html')
    print("The website has been successfully brought.")
else:
    print("Failed to fetch the website.")
```

```python
game_team = soup.find('section',{'id':'game-team'})
table = game_team.find('table',{'class':'sidearm-table'})
table
```

```python
headers_row = table.find('thead',{'class':'sidearm-primary'})
headers = headers_row.find_all('th')

columns_gbg = [column.text.strip() for column in headers]
columns_gbg.append('Link')
columns_gbg
```

```python
df_game_by_game = pd.DataFrame(columns = columns_gbg)
df_game_by_game
```

```python
table_body = table.find('tbody')
match_rows = table_body.find_all('tr')
match_rows
```

```python
for row in match_rows:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    link = row_data[0].find('a')['href']
    id = link.split('=')[1].split('&')[0]
    individual_row_data.append(path+id)

    print(individual_row_data)

    temp_df = pd.DataFrame([individual_row_data],columns=columns_gbg)
    df_game_by_game = pd.concat([df_game_by_game, temp_df], ignore_index=True)

df_game_by_game.to_csv('game_by_game.csv',index=False)
```

```python
temp_url = 'https://loyolagreyhounds.com/sports/womens-basketball/stats/2022-23/niagara/boxscore/15531'
response = requests.get(temp_url)
soup = BeautifulSoup(response.content,'html')
print(f'Successfully received {link}')
```

```python
extra_header= ['team','opponent']
```

```python
main = soup.find('main',class_='main-content-placeholder',id='main-content')
page = main.find('div',class_='sidearm-responsive-tabs')
panels = page.find('section',id='box-score')
team_panels = panels.find_all('section',class_='panel')

oppoment = team_panels[0] 
own= team_panels[1]

oppoment_team_name = oppoment.find('h3').text
own_team_name = own.find('h3').text


oppoment_team_name,own_team_name
```

```python
table = oppoment.find_all('table',class_="sidearm-table")[0] #for headers 
thead = table.find('thead')
row = thead.find_all('th')

columns_box_score = [data.text.strip() for data in row]
columns_box_score = extra_header + columns_box_score
columns_box_score
```

```python
df_box_score = pd.DataFrame(columns = columns_box_score)
df_box_score
```

```python
data_list = []

for i,(team) in enumerate(team_panels):
    table = team.find_all('table',class_="sidearm-table")[0]
    tbody = table.find('tbody')
    players = tbody.find_all('tr')
    
    team_name = team_panels[-i].find('h3').text
    oppoment_name = team_panels[-i+1].find('h3').text
    
    pre_data = [team_name,oppoment_name]
    
    for player in players:
        row = player.find_all('td')
        data = [data.text.strip() for data in row]
        
        data = pre_data + data

        temp_df = pd.DataFrame([data],columns=columns_box_score)
        df_box_score = pd.concat([df_box_score, temp_df], ignore_index=True)

    
df_box_score
```
