```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df_game_by_game = pd.read_csv('../game_by_game_table/all_game_by_game.csv')
df_game_by_game
```

```python
front_header = ['team','opponent']
back_header = ['date','time','site']
```

```python
data_list = []

for link in df_game_by_game['Link']:
    response = requests.get(link)
    soup = BeautifulSoup(response.content,'html')
    print(f'Successfully received {link}')
    
    main = soup.find('main',class_='main-content-placeholder',id='main-content')
    page = main.find('div',class_='sidearm-responsive-tabs')
    panels = page.find('section',id='box-score')
    team_panels = panels.find_all('section',class_='panel')
    
    info_panels = panels.find('header',class_='panel')
    #mathc_details = info_panels.find('div',class_='large-4')
    match_details = info_panels.find('aside',class_='game-details')
    date = match_details.find_all('dd')[0].text
    time = match_details.find_all('dd')[1].text
    site = match_details.find_all('dd')[3].text
    back_data = [date,time,site]
    
    #oppoment = team_panels[0] 
    #own= team_panels[1]

    for i,(team) in enumerate(team_panels):
        table = team.find_all('table',class_="sidearm-table")[0]
        tbody = table.find('tbody')
        players = tbody.find_all('tr')

        team_name = team_panels[-i].find('h3').text
        oppoment_name = team_panels[-i+1].find('h3').text

        front_data = [team_name,oppoment_name]

        for player in players:
            row = player.find_all('td')
            data = [data.text.strip() for data in row]

            data = front_data + data + back_data
            data_list.append(data)

            #temp_df = pd.DataFrame([data],columns=columns_box_score)
            #df_box_score = pd.concat([df_box_score, temp_df], ignore_index=True)


thead = table.find('thead')
row = thead.find_all('th')
    
columns_box_score = [data.text.strip() for data in row]
columns_box_score = front_header + columns_box_score + back_header
df_box_score = pd.DataFrame(data_list,columns = columns_box_score)
```

```python
df_box_score.to_csv('all_box_score.csv',index=False)
```
