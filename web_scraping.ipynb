```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://loyolagreyhounds.com/sports/womens-basketball/roster?view=2'
response = requests.get(url)

if response.status_code == 200:
    html_content = response.content
else:
    print("Failed to fetch the website.")
```

```python
soup = BeautifulSoup(response.text,'html')
```

```python
roster_table = soup.find_all('table')[-2]
coach_table = soup.find_all('table')[-1]
roster_table
```

```python
roster_tittles = roster_table.find_all('th')
roster_tittles
```

```python
roster_table_tittles = [tittle.text for tittle in roster_tittles]
roster_table_tittles
```

```python
df = pd.DataFrame(columns = roster_table_tittles)
df
```

```python
player_row = roster_table.find_all('tr')
player_row
```

```python
pd.concat
```

```python
data_list = []

for row in player_row:
    try:
        row_data = row.find_all('td')
        individual_row_data = [data.text.strip() for data in row_data]
        print(individual_row_data)

        temp_df = pd.DataFrame([individual_row_data],columns=roster_table_tittles)
        df = pd.concat([df, temp_df], ignore_index=True)
    except ValueError:
        pass
```
